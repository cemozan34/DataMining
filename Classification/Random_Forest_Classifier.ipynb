```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
columns = ["trustLevel","totalScannedItems","lineItemVoids","scansWithoutRegistration","valuePerSecond","lineItemVoidsPerPosition","fraud"]
df_train = pd.read_csv("train.csv", delimiter="|")
df_test = pd.read_csv("test.csv", delimiter="|")
```

```python
df_train['totalScannedItems'] = df_train.totalScanTimeInSeconds * df_train.scannedLineItemsPerSecond
df_train = df_train.drop(['totalScanTimeInSeconds','scannedLineItemsPerSecond'],axis=1)

df_test['totalScannedItems'] = df_test.totalScanTimeInSeconds * df_test.scannedLineItemsPerSecond
df_test = df_test.drop(['totalScanTimeInSeconds','scannedLineItemsPerSecond'],axis=1)
```

```python
df_train.head()
```

```python
x = {
    1:5,
    2:4,
    3:3,
    4:2,
    5:1,
    6:0
    }

df_train["trustLevel"] = df_train["trustLevel"].map(x)
df_test["trustLevel"] = df_test["trustLevel"].map(x)
```

```python
X = np.array(df_train.loc[:,["trustLevel","totalScannedItems","lineItemVoids","scansWithoutRegistration","valuePerSecond","lineItemVoidsPerPosition"]
])
y = np.array(df_train["fraud"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
accuracy_score(y_test,pred)
```

```python
print("Accuracy score: ")
print(accuracy_score(y_test,pred))
print("Confusion Matrix: ")
print(confusion_matrix(y_test,pred))
```

```python
def total_cost(confusion_matrix):
    total_cost = 0
    total_cost += (confusion_matrix[1][0] * -5)
    total_cost += (confusion_matrix[1][1] * 5)
    total_cost += (confusion_matrix[0][1] * -25)
    return total_cost
```

```python
total_cost(confusion_matrix(y_test,pred))
```

```python
print('Index\tPredicted\t\tActual')
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        print(i,'\t',pred[i],'\t\t\t',y_test[i],'******')
```
